Sort an Array of 0s, 1s, and 2s

```python
class Solution:
    def sort012(self, arr):
        # code here
        zeroes = arr.count(0)
        ones = arr.count(1)
        twos = arr.count(2)
        
        arr.clear()
        for x in range(zeroes):
            arr.append(0)
        for x in range(ones):
            arr.append(1)
        for x in range(twos):
            arr.append(2)
            
        # this works, but i don't like the fact that it only works for 0s, 1s, and 2s.
        # i need to relearn how to do sorting algorithms, apparently.
```

Binary Search

```python
class Solution:
    def binarysearch(self, arr, k):
        # Code Here
        low = 0
        high = len(arr) - 1
        
        # index of k. will be the first instance of k by the end of the program
        k_index = -1
        
        found_an_instance = False
        
        # find an instance of k
        while (low <= high) and not found_an_instance:
            mid = (low + high) // 2
            
            # if an instance is found, exit
            if k == arr[mid]:
                k_index = mid
                found_an_instance = True
            # if k is less than arr[mid], search to the left of arr[mid]
            elif k < arr[mid]:
                high = mid - 1
            # if k is more than arr[mid], search to the right of arr[mid]
            else:
                low = mid + 1
                
        # if k is found, go to lower index until first instance of k
        if found_an_instance:
            while k == arr[mid]:
                k_index = mid
                mid -= 1
                
        return k_index
```
